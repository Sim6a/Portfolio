# Восстановление золота из руды

**Задача**

Подготовить прототип модели машинного обучения для «Цифры». Компания разрабатывает решения для эффективной работы промышленных предприятий.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Необходимо использовать данные с параметрами добычи и очистки. 

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Необходимо:

1. Подготовить данные;
2. Провести исследовательский анализ данных;
3. Построить и обучить модель.

**Описание данных**

- Данные индексируются датой и временем получения информации (признак date);
- Соседние по времени параметры часто похожи;
- Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже;
- Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей;
- Также в тестовом наборе нет целевых признаков;
- Исходный датасет содержит обучающую и тестовую выборки со всеми признаками;

В нашем распоряжении сырые данные: их просто выгрузили из хранилища. Прежде чем приступить к построению модели, необходимо проверить их на корректность.

**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики
Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка рассчетов

Необходимо проверить, что что эффективность обогащения рассчитана правильно.

Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между полученными расчётами и значением признака.

**Формула:** Recovery = C * (F - T) / F * (C - T)

Где:

- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

```python
recovery = df_train['rougher.output.recovery']
f = df_train['rougher.input.feed_au']
c = df_train['rougher.output.concentrate_au']
t = df_train['rougher.output.tail_au']

rec = (c * (f - t)) / (f * (c - t)) * 100
rec = pd.Series(rec)
rec.head()
```

**Найдем среднее абсолютное отклонение МАЕ**

```python
print(mean_absolute_error(recovery, rec))
```

Значение MAE близко к нулю, значит, изначальные значения признака rougher.output.recovery рассчитаны верно.

## Анализ данных

### Посмотрим, как меняется концентрация металлов на различных этапах

```python
# выбираем данные для визуализации
def select(df):
    au = df[['rougher.input.feed_au',
             'rougher.output.concentrate_au', 
             'primary_cleaner.output.concentrate_au',
             'final.output.concentrate_au']]
    ag = df[['rougher.input.feed_ag',
             'rougher.output.concentrate_ag', 
             'primary_cleaner.output.concentrate_ag',
             'final.output.concentrate_ag']]
    pb = df[['rougher.input.feed_pb',
             'rougher.output.concentrate_pb',
             'primary_cleaner.output.concentrate_pb',
             'final.output.concentrate_pb']]
    return au, ag, pb

def vis(df, title):    
    plt.figure(figsize=(15,8))
    ax = sns.histplot(df)
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('Концентрация металла в %', fontsize=16)
    ax.set_ylabel('Количество наблюдений')
    ax.legend(['Финальный концентрат', 'После первичной очистки', 'После флотации', 'Начальное сырье'])
```

```python
au, ag, pb = select(df)
vis(au, 'Концентрация золота на разных этапах') 
vis(ag, 'Концентрация серебра на разных этапах')  
vis(pb, 'Концентрация свинца на разных этапах')  
```

Присутствуют всплески околонулевых значения, которые, вероятнее всего, являются выбросами, их необходимо исключить.

```python
def clean(df):
    df = df.loc[((df['rougher.input.feed_au'] > 1) 
                 & (df['rougher.output.concentrate_au'] > 1)
                 & (df['primary_cleaner.output.concentrate_au'] > 1)
                 & (df['final.output.concentrate_au'] > 1)
                 & (df['rougher.input.feed_ag'] > 1)
                 & (df['rougher.output.concentrate_ag'] > 1)
                 & (df['primary_cleaner.output.concentrate_ag'] > 1)
                 & (df['final.output.concentrate_ag'] > 1)
                 & (df['rougher.input.feed_pb'] > 1)
                 & (df['rougher.output.concentrate_pb'] > 1)
                 & (df['primary_cleaner.output.concentrate_pb'] > 1)
                 & (df['final.output.concentrate_pb'] > 1))] 
    return df

df = clean(df)
```

```python
au, ag, pb = select(df)
vis(au, 'Концентрация золота на разных этапах') 
vis(ag, 'Концентрация серебра на разных этапах')  
vis(pb, 'Концентрация свинца на разных этапах')  
```

## Модель

### Определяем признаки и целевые признаки

Исключим из обучающей выборки признаки, которые отсутствуют в тестовой, т.к. они описывают результат какой-либо стадии, и не могут использоваться для обучения.

```python
# в diff хранится разница признаков между тестовой и обучающей выборках
diff.remove('rougher.output.recovery')
diff.remove('final.output.recovery')
df_train = df_train.drop(diff, axis=1)

# удалим столбцы с суммарными рассчетами
droplist = ['rougher.input.feed_sum', 'rougher.output.concentrate_sum',
            'primary_cleaner.output.concentrate_sum', 'final.output.concentrate_sum']
df = df.drop(droplist, axis=1)
```

#### Признаки для оценки эффективности очистки чернового концентрата

Нам нужно отобрать все признаки, в названии которых присутствует 'rougher'

```python
rougher = list(filter(lambda x: 'rougher' in x, df_train.columns))
rougher.remove('rougher.output.recovery')

rougher_features_train = df_train[rougher]
rougher_features_test = df_test[rougher]

rougher_target_train = df_train['rougher.output.recovery']

# выберем значения целевого признака тестовой выборки по дате
d = df_test['date']
rougher_target_test = df.query('date in @d')['rougher.output.recovery']
rougher_target_test.index = rougher_features_test.index

print(rougher_features_train.shape)
print(rougher_target_train.shape)
print(rougher_features_test.shape)
print(rougher_target_test.shape)
```

#### Признаки для оценки эффективности очистки финального концентрата

Возьмем все признаки, кроме даты, т.к. все они оказывают влияние на финальный результат.

```python
final = list(df_train.columns)
final.remove('date')
final.remove('rougher.output.recovery')
final.remove('final.output.recovery')

final_features_train = df_train[final]
final_features_test = df_test[final]

final_target_train = df_train['final.output.recovery']

# выберем значения целевого признака тестовой выборки по дате
d = df_test['date']
final_target_test = df.query('date in @d')['final.output.recovery']
final_target_test.index = final_features_test.index

print(final_features_train.shape)
print(final_target_train.shape)
print(final_features_test.shape)
print(final_target_test.shape)
```

### sMAPE

Для решения задачи введём новую метрику качества — sMAPE.

```python
def smape(target, predictions):
    result = abs(target - predictions) / ((abs(target) + abs(predictions)) / 2) * 100
    return result.mean()

score = make_scorer(smape, greater_is_better=False)
```

Напишем функцию для подсчета итогового sMAPE 

```python
def final_smape(rough, final):
    result = 0.25 * rough + 0.75 * final
    return result.mean()
```

### Выполним масштабирование

```python
scaler = StandardScaler()

scaler.fit(rougher_features_train)
rougher_features_train = scaler.transform(rougher_features_train)
rougher_features_test = scaler.transform(rougher_features_test)

scaler.fit(final_features_train)
final_features_train = scaler.transform(final_features_train)
final_features_test = scaler.transform(final_features_test)
```

### Выбираем лучшую модель для оценки эффективности обогащения чернового концентрата 

#### Линейная регрессия

Подбираем параметры

```python
%%time

params_linreg = {}

model_linreg_rough = LinearRegression(copy_X=True, fit_intercept=True, positive=False)
grid = GridSearchCV(model_linreg_rough, params_linreg, scoring=score, cv=3, n_jobs=-1)
grid.fit(rougher_features_train, rougher_target_train)
#print('Параметры:', grid.best_params_)
smape_rough_linreg = grid.best_score_
print('sMAPE:', -smape_rough_linreg)
```

#### Решающее дерево

```python
%%time

params_tree = {
          'max_depth': range (1, 6),
          'min_samples_leaf': range (1, 20, 2),
          'min_samples_split': range (2, 20, 2) 
              }

model_tree_rough = DecisionTreeRegressor(random_state=state)
grid = GridSearchCV(model_tree_rough, params_tree, scoring=score, cv=3, n_jobs=-1)
grid.fit(rougher_features_train, rougher_target_train)
print('Параметры:', grid.best_params_)
smape_rough_tree = grid.best_score_
print('sMAPE:', -smape_rough_tree)
```

#### Случайный лес

```python
%%time

params_forest = {
          'n_estimators': range (10, 30, 10),  
          'max_depth': range (1, 6),
          'min_samples_leaf': range (1, 22, 4),
          'min_samples_split': range (2, 14, 4) 
                }

model_forest_rough = RandomForestRegressor(random_state=state)
grid = GridSearchCV(model_forest_rough, params_forest, scoring=score, cv=3, n_jobs=-1)
grid.fit(rougher_features_train, rougher_target_train)
print('Параметры:', grid.best_params_)
smape_rough_forest = grid.best_score_
print('sMAPE:', -smape_rough_forest)
```

### Выбираем лучшую модель для оценки эффективности обогащения финального концентрата 

#### Линейная регрессия

```python
%%time

model_linreg_final = LinearRegression(copy_X=True, fit_intercept=True, positive=False)
grid = GridSearchCV(model_linreg_final, params_linreg, scoring=score, cv=3, n_jobs=-1)
grid.fit(final_features_train, final_target_train)
#print('Параметры:', grid.best_params_)
smape_final_linreg = grid.best_score_
print('sMAPE:', -smape_final_linreg)
```

```python
final_smape_linreg = final_smape(smape_rough_linreg, smape_final_linreg)
print('Итоговое sMAPE линейной регрессии:', -final_smape_linreg)
```

#### Решающее дерево

```python
%%time

model_tree_final = DecisionTreeRegressor(random_state=state)
grid = GridSearchCV(model_tree_final, params_tree, scoring=score, cv=3, n_jobs=-1)
grid.fit(final_features_train, final_target_train)
print('Параметры:', grid.best_params_)
smape_final_tree = grid.best_score_
print('sMAPE:', -smape_final_tree)
```

```python
final_smape_tree = final_smape(smape_rough_tree, smape_final_tree)
print('Итоговое sMAPE решающего дерева:', -final_smape_tree)
```

#### Случайный лес

```python
%%time

model_forest_final = RandomForestRegressor(random_state=state)
grid = GridSearchCV(model_forest_final, params_forest, scoring=score, cv=3, n_jobs=-1)
grid.fit(final_features_train, final_target_train)
print('Параметры:', grid.best_params_)
smape_final_forest = grid.best_score_
print('sMAPE:', -smape_final_forest)
```

```python
final_smape_forest = final_smape(smape_rough_forest, smape_final_forest)
print('Итоговое sMAPE случайного леса:', -final_smape_forest)
```

Для оценки финального концентрата, а так же по результатам итогового sMAPE, лучше всего показала себя модель случайного леса.

### Проверка на адекватность

Сравним модели случайного леса обоих этапов с константной моделью, которая будет предсказывать среднее.

```python
dummy = DummyRegressor(strategy='mean')
dummy.fit(rougher_features_train, rougher_target_train)
dummy_pred = dummy.predict(rougher_features_test)
smape_rough_dummy = smape(rougher_target_test, dummy_pred)
print('Dummy sMAPE rough', smape_rough_dummy.mean())

dummy.fit(final_features_train, final_target_train)
dummy_pred = dummy.predict(final_features_test)
smape_final_dummy = smape(final_target_test, dummy_pred)
print('Dummy sMAPE final', smape_final_dummy.mean())
final_smape_dummy = final_smape(smape_rough_dummy, smape_final_dummy)
print('Итоговое sMAPE для константной модели:', final_smape_dummy)
```

```python
model_results = pd.DataFrame({'Модель' : ['Константная', 'Линейная регрессия', 'Решающее дерево', 'Случайный лес'],
                              'sMAPE флотации' :[smape_rough_dummy,
                                            -smape_rough_linreg,
                                            -smape_rough_tree,
                                            -smape_rough_forest],
                             'sMAPE финального концентрата' : [smape_final_dummy,
                                            -smape_final_linreg,
                                            -smape_final_tree,
                                            -smape_final_forest],
                             'Итоговое sMAPE' : [final_smape_dummy,
                                            -final_smape_linreg,
                                            -final_smape_tree,
                                            -final_smape_forest]
                             })

display(model_results)
```

Видим, что у константной модели резульаты хуже. Все наши модели проходят проверку на адекватность. Лучшая модель - случайный лес.

### Тестирование

Выполним тестирование полученных моделей и получим финальное sMAPE для тестовой выборки.

```python
# для флотации
model_forest_rough = RandomForestRegressor(max_depth=5, 
                                           min_samples_leaf=13,
                                           min_samples_split=2,
                                           n_estimators=20,
                                           random_state=state)
model_forest_rough.fit(rougher_features_train, rougher_target_train)
preds_test_rough = model_forest_rough.predict(rougher_features_test)
smape_rough_test = smape(rougher_target_test, preds_test_rough)
print('Test rough sMAPE', smape_rough_test.mean())

# дя финального концентрата
model_forest_final = RandomForestRegressor(max_depth=5, 
                                           min_samples_leaf=13,
                                           min_samples_split=2,
                                           n_estimators=10,
                                           random_state=state)
model_forest_final.fit(final_features_train, final_target_train)
preds_test_final = model_forest_final.predict(final_features_test)
smape_final_test = smape(final_target_test, preds_test_final)
print('Test final sMAPE', smape_final_test.mean())

smape_total = final_smape(smape_rough_test, smape_final_test)
print('Итоговое sMAPE лучших моделей после тестирования:', smape_total)
```
