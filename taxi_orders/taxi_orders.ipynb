#  Прогнозирование заказов такси

Компания собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Нужно построить модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

**План работы:**

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.

## Подготовка

### Импортируем библиотеки

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

from statsmodels.tsa.seasonal import seasonal_decompose

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, make_scorer

state = np.random.RandomState(12345)
```

### Загружаем и проверяем датасет

```python
path1 = 'taxi.csv'
path2 = '/datasets/taxi.csv'

def read_file(path1, path2):
    if os.path.exists(path1):
        df = pd.read_csv(path1, parse_dates = [0])
    elif os.path.exists(path2):
        df = pd.read_csv(path2, parse_dates = [0])
    else:
        print('File not found')
    return df

df = read_file(path1, path2)
```

```python
display(df.info())
display(df.head())
```

Данные загружены корректно, пропусков нет.

### Ресемплирование
Зададим в качестве индекса столбец с датой, отсортируем по возрастанию и выполним ресемплирование данных по одному часу.

```python
df.set_index('datetime', drop=True, inplace=True)
df.sort_index(ascending=True, inplace=True)
df = df.resample('1H').sum()
df.head()
```

```python
df['rolling_mean_d'] = df['num_orders'].rolling(24).mean()
df['rolling_mean_w'] = df['num_orders'].rolling(24*7).mean()

ax = df.plot(figsize = [10, 6], title = 'Число заказов', xlabel = 'Месяц' )
ax.legend(['Число заказов', 'Среднее в день', 'Среднее в неделю']);

df = df.drop(['rolling_mean_d', 'rolling_mean_w'], axis=1)
```

Видим тренд на увеличение числа заказов. Заметно, что ближе к концу августа, рост усилился - вероятнее всего, это связано с погодой и сменой сезона.

Рассмотрим более детально один день.

```python
decomposed.seasonal['2018-05-20':'2018-05-21'].plot(ax=plt.gca(), figsize= [10, 6]);
```

```python
df1 = df.resample('1D').sum()
decomposed1 = seasonal_decompose(df1)
ax = decomposed1.seasonal['2018-05-01':'2018-05-15'].plot(ax=plt.gca(), figsize= [10, 6], \
                                                          title='Сезонность по неделям', xlabel='Дата');
```

Так же видим недельную сезонность - постепенный рост в течении первых трёх дней, затем спад ближе к выходным, и снова резкий скачок с началом новой недели.

## Обучение

### Подготовка данных

#### Подготовка признаков

Добавим в датасет столбцы с месяцем, днем и часом заказа, скользящее среднее и отстающие значения.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
make_features(df, 24, 168)
display(df.head())
display(df.shape)
```

#### Разобьем данные на обучающую и тестовую выборки

```python
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()
display(train.shape)
display(test.shape)
```

#### Определим признаки и целевой признак

```python
features_train = train.drop('num_orders', axis=1)
features_test = test.drop('num_orders', axis=1)
target_train = train['num_orders']
target_test = test['num_orders']
```

#### Выполним масштабирование признаков

```python
numeric = df.columns.drop('num_orders')

scaler = StandardScaler()
scaler.fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

### Метрика

```python
def rmse(answ, pred):
    return (mean_squared_error(answ, pred) ** 0.5)

score = make_scorer(rmse, greater_is_better=False)
```

### Обучение

```python
tscv = TimeSeriesSplit(n_splits=5)
```

#### Линейная регрессия

```python
%%time

model_lin_reg = LinearRegression(fit_intercept=False, normalize=True, copy_X=False)

grid_lr = GridSearchCV(model_lin_reg, param_grid={}, scoring=score, cv=tscv, n_jobs=-1)
grid_lr.fit(features_train, target_train)
print('RMSE линейной регрессии', -grid_lr.best_score_)
```

#### Решающее дерево

```python
%%time

params_tree = {
          'max_depth': range (1, 6),
          'min_samples_leaf': range (1, 20, 2),
          'min_samples_split': range (2, 20, 2) 
              }

model_tree = DecisionTreeRegressor(random_state=state)
grid_tree = GridSearchCV(model_tree, params_tree, scoring=score, cv=tscv, n_jobs=-1)
grid_tree.fit(features_train, target_train)
print('Параметры:', grid_tree.best_params_)
print('RMSE Решающего дерева:', -grid_tree.best_score_)
```

#### LightGBM

```python
%%time

params_lgbm = {
                'learning_rate': [0.1, 0.2],
                'n_estimators': [100, 250],
                'max_depth': [6, 12],
                'num_leaves': [6, 12]
              }

model_lgbm = LGBMRegressor(random_state = state)
grid_lgbm = GridSearchCV(model_lgbm, params_lgbm, scoring=score, cv=tscv, n_jobs=-1)
grid_lgbm.fit(features_train, target_train)
print('Параметры:', grid_lgbm.best_params_)
print('RMSE LightGBM:', -grid_lgbm.best_score_)
```

#### CatBoost

```python
%%time

params_cbr = {
                'depth'         : [6, 8, 10],
                'learning_rate' : [0.01, 0.05, 0.1],
                'iterations'    : [30, 50, 150]
              }

model_cbr = CatBoostRegressor()
grid_cbr = GridSearchCV(model_cbr, params_cbr, scoring=score, cv=tscv, n_jobs=-1)
grid_cbr.fit(features_train, target_train)
print('Параметры:', grid_cbr.best_params_)
print('RMSE CatBoostRegressor:', -grid_cbr.best_score_)
```

### Анализ моделей

```python
results = pd.DataFrame({
    'Модель':          ['Linear Regression', 'Decision Tree', 'LightGBM', 'CatBoost Regressor'],
    'RMSE':            [-grid_lr.best_score_,
                       -grid_tree.best_score_,
                       -grid_lgbm.best_score_,
                       -grid_cbr.best_score_]                      
                                                
})
display(results)
```

### Промежуточный вывод

- добавили в датасет столбцы с месяцем, днем и часом заказа, скользящее среднее и отстающие значения;
- разбили данные на обучающую и тестовую выборки;
- определили признаки и целевой признак;
- задали оценщик;
- обучили несколько моделей при помощи GridSearchCV;
- лучше всех показала себя модель CatBoost Regressor;

## Тестирование

```python
model = CatBoostRegressor(**grid_cbr.best_params_)
model.fit(features_train, target_train)
preds = model.predict(features_test)
print('RMSE после тестирования', rmse(target_test, preds))
```
