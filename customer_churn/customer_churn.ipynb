# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Подготовка данных

### Импортируем библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.dummy import DummyClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

from sklearn.utils import shuffle
```

### Загружаем датасет

```python
df = pd.read_csv('Churn.csv')
display(df.head())
print(df.info())
```

### Преобразуем типы для экономии памяти

```python
numeric_int = ['RowNumber', 'CustomerId', 'CreditScore', 'Age', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Exited']
for i in numeric_int:
    df[i] = pd.to_numeric(df[i], downcast='integer')

numeric_float = ['Tenure', 'Balance', 'EstimatedSalary']
for i in numeric_float:
    df[i] = pd.to_numeric(df[i], downcast='float')
    
df.info()
```

### Удаляем пропуски

Пропуски составляют около 10% от набора данных.

```python
df = df.dropna().reset_index(drop=True)
df.info()
```

### Определим признаки и целевой признак

Столбцы RowNumber, CustomerId и Surname нам не нужны, исключим их.

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

### Выполним прямое кодирование категориальных столбцов

Наши данные содержат как количественные, так и категориальные признаки. Чтобы модели могли работать с последними, выполним прямое кодирование.

```python
df_ohe = pd.get_dummies(df, drop_first=True)

# определяем признаки
target = df_ohe['Exited']
features = df_ohe.drop('Exited', axis=1)
```

### Разделим данные на обучающую, валидационную и тестовую выборки в соотношении 3:1:1

```python
features_train, features_temp, target_train, target_temp = train_test_split(
    features, target, test_size=0.6, random_state=12345)

features_valid, features_test, target_valid, target_test = train_test_split(
    features_temp, target_temp, test_size=0.5, random_state=12345)

print('train:', features_train.shape, target_train.shape)
print('valid:', features_valid.shape, target_valid.shape)
print('test:', features_test.shape, target_test.shape)
```

### Выполним масштабирование численных признаков

Чтобы модели считали все признаки одинаково значимыми, применим стандартизацию.

```python
numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']

pd.options.mode.chained_assignment = None

scaler = StandardScaler()
scaler.fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

## Исследование задачи

### Решающее дерево

#### Подбор гиперпараметров

Используем GridSearch с параметром scoring = 'f1'

```python
%%time
print()

params = {
          'max_depth': range (1, 15),
          'min_samples_leaf': range (1, 20),
          'min_samples_split': range (2, 20, 2) 
         }

model_tree = DecisionTreeClassifier(random_state=12345)
model_tree.fit(features_train, target_train)

grid = GridSearchCV(model_tree, params, scoring='f1', cv=5)
grid.fit(features_train, target_train)
print(grid.best_params_)
```

#### Проверка модели на адекватность

Выведем Accuracy, F1-меру и AUC ROC, сравним с константной моделью.

```python
model_tree = DecisionTreeClassifier(random_state=12345, max_depth=9, min_samples_leaf=17, min_samples_split=2)
model_tree.fit(features_train, target_train)
predictions_valid = model_tree.predict(features_valid)
print('Accuracy:', accuracy_score(target_valid, predictions_valid))
tree_f1_default = f1_score(target_valid, predictions_valid)
print('F1:', tree_f1_default)
```

```python
dummy = DummyClassifier(random_state=12345, strategy='most_frequent')
dummy.fit(features_train, target_train)
dummy_pred_valid = dummy.predict(features_valid)
print('Accuracy dummy:', accuracy_score(target_valid, dummy_pred_valid))
```

Наша модель проходит проверку на адекватность, но даже у константной модели 80% доля правильных ответов.

#### ROC-кривая

```python
probabilities_valid = model_tree.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)

plt.figure()
plt.plot(fpr, tpr)

# ROC-кривая случайной модели
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()
tree_roc_auc_default = roc_auc_score(target_valid, predictions_valid)
print('AUC ROC:', tree_roc_auc_default)
```

## Борьба с дисбалансом

### Проверяем количество признаков со значениями 0 и 1

```python
features_zeros = features_train[target_train == 0]
features_ones = features_train[target_train == 1]
target_zeros = target_train[target_train == 0]
target_ones = target_train[target_train == 1]

print(features_zeros.shape)
print(features_ones.shape)
print(target_zeros.shape)
print(target_ones.shape)
```

Перевес идет в сторону отрицательных классов.

### Применим upsampling - увеличение выборки

Будем увеличивать число положительных признаков в 4 раза, чтобы уравнять с отрицательными.

```python
features_upsampled = pd.concat([features_zeros] + [features_ones] * 4)
target_upsampled = pd.concat([target_zeros] + [target_ones] * 4)

features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)
```

#### Проверим модель решающего дерева

```python
model_tree = DecisionTreeClassifier(random_state=12345, max_depth=9, min_samples_leaf=17, min_samples_split=2)
model_tree.fit(features_upsampled, target_upsampled)

predictions_valid = model_tree.predict(features_valid)
tree_f1_score_upsample = f1_score(target_valid, predictions_valid)
print('F1:', tree_f1_score_upsample)
```

```python
probabilities_valid = model_tree.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)

plt.figure()
plt.plot(fpr, tpr)

# ROC-кривая случайной модели
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()

tree_auc_roc_score_upsample = roc_auc_score(target_valid, predictions_valid)
print('AUC ROC:', tree_auc_roc_score_upsample)
```
