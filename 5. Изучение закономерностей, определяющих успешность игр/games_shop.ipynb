# Интернет-магазин видеоигр "Стримчик"

Нам доступны исторические данные о продажах игр до 2016 года, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Описание данных

Датасет - games.csv:

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Шаг - предобработка данных.

### Приводим названия столбцов и категориальные значения к нижнему регистру:

```python
df.columns = df.columns.str.lower()
df['name'] = df['name'].str.lower()
df['platform'] = df['platform'].str.lower()
df['rating'] = df['rating'].str.lower()
```

### Изучаем столбцы, обрабатываем пропуски и преобразуем типы данных:

##### year_of_release

```python
display(df['year_of_release'].describe())
display(df['year_of_release'].unique())
```

Аномальных значений нет, приводим к целому типу.

```python
df['year_of_release'] = pd.to_numeric(df['year_of_release'], downcast='integer')
```

##### na_sales

```python
display(df['na_sales'].describe())
display(df['na_sales'].value_counts())
```

Много нулевых значений. Но игры могли и не продаваться в этом регионе. Оставим это и учтём в дальнейшем анализе. Приводим к типу float32.

```python
df['na_sales'] = pd.to_numeric(df['na_sales'], downcast='float')
```

##### eu_sales

```python
display(df['eu_sales'].describe())
display(df['eu_sales'].value_counts())
```

Так же много нулевых значений. Но игры могли и не продаваться в этом регионе. Оставим это и учтём в дальнейшем анализе. Приводим к типу float32.

```python
df['eu_sales'] = pd.to_numeric(df['eu_sales'], downcast='float')
```

##### jp_sales

```python
display(df['jp_sales'].describe())
display(df['jp_sales'].value_counts())
```

Еще больше нулевых значений. Япония - особенный регион, ориентированный на внутренний рынок. Многие игры туда не доезжают, многие - доступны только на внутреннем рынке. Оставим это и учтём в дальнейшем анализе. Приводим к типу float32.

```python
df['jp_sales'] = pd.to_numeric(df['jp_sales'], downcast='float')
```

##### other_sales

```python
display(df['other_sales'].describe())
display(df['other_sales'].value_counts())
```

Как и в предыдущих случаях - оставляем нулевые значения. Приводим к float32.

```python
df['other_sales'] = pd.to_numeric(df['other_sales'], downcast='float')
```

##### critic_score

```python
display(df['critic_score'].describe())
display(df['critic_score'].value_counts())
display(df['critic_score'].unique())
```

Видим большое количество пропусков в этом столбце. Некоторым играм оценку могли не поставить, это нормально. Минимальное значение 13, значит, можно заменить пропуски на 0 и, в дальнейшем, рассматривать это как отсутсвие оценки.

Приведем к типу int.

```python
df['critic_score'] = df['critic_score'].fillna(0)
df['critic_score'] = pd.to_numeric(df['critic_score'], downcast='integer')
```

##### user_score

```python
display(df['user_score'].describe())
display(df['user_score'].value_counts())
display(df['user_score'].unique())
```

Присутствуют как пропуски, так и значение tbd, что означает To Be Determined - то есть, игре еще не поставили оценку. Это может быть связано с тем, что аггрегатору нехватило пользовательских оценок, чтобы рассчитать рейтинг. Либо издатель отказался публиковать информацию о рейтинге. В любом случае, это означает пропуск.

Минимальная оценка 0 всего у одной игры. Заменим пропуски и tbd на 0 и будем считать это отсутствием оценки. Так же домножим результат на 10 и приведем к целому типу. Это может пригодиться в дальнейшем для сравнения с рейтингом критиков.

```python
df['user_score'] = df['user_score'].fillna(0)
df.loc[df['user_score'] == 'tbd', 'user_score'] = 0
df['user_score'] = df['user_score'].astype(float)
df['user_score'] = df['user_score'] * 10
df['user_score'] = pd.to_numeric(df['user_score'], downcast='integer')
df['user_score'].unique()
```

##### rating

```python
display(df['rating'].describe())
display(df['rating'].value_counts())
display(df['rating'].unique())
```

Со значениями все в порядке. Пропуски можно заменить на tbd, это будет означать, что рейтинг не выставлен.

```python
df['rating'] = df['rating'].fillna('tbd')
```

### Построим диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами на самой популярной платформе PS4.

Нулевых оценок пользователей и критиков в исходных данных не было. Нулями мы обозначили отсутствие оценок. Исключим нулевые значения user_score и critic_score из визуализации.

```python
df_ps4_users = df[(df['platform'] == 'ps4') & (df['user_score'] != 0)]

sns.relplot(x='user_score', y='sales_sum', data=df_ps4_users)
plt.title('Зависимость между оценками пользователей и продажами')
plt.xlabel('Оценка')
plt.ylabel('Продажи');
print('Коэффициент корреляции между оценками пользователей и продажами: ', \
      round(df_ps4_users['user_score'].corr(df_ps4_users['sales_sum']), 5)
     )
```

Можно сказать, что продажи практичеси не зависят от оценок пользователей. Корреляция крайне мала, диаграмма показывает, что даже у высокооцененных пользователями игр могут быть очень маленькие продажи.

```python
# нули - это отсутсвтие оценок, их не учитываем
df_ps4_crit = df[(df['platform'] == 'ps4') & (df['critic_score'] != 0)]

sns.relplot(x='critic_score', y='sales_sum', data=df_ps4_crit)
plt.title('Зависимость между оценками критиков и продажами')
plt.xlabel('Оценка')
plt.ylabel('Продажи');
print('Коэффициент корреляции между оценками пользователей и продажами: ', \
      round(df_ps4_crit['critic_score'].corr(df_ps4_crit['sales_sum']), 2)
     )
```

Зависимость между продажами и оценками критиков намного выше, но все же не слишком большая. Можно утверждать, что при покупке игр пользователи обращают внимание на мнения авторитетных издательств, однако в большей степени на продажи влияет что-то еще, возможно, рекламная кампания.

## Шаг - проверка гипотез.

### Часть 1

Используем гипотезу о равенстве средних двух генеральных совокупностей, проверяем с помощью st.ttest_ind( )

H0: средний пользовательский рейтинг XOne равен среднему пользовательскому рейтингу PC;

H1: средний пользовательский рейтинг XOne больше среднего пользовательского рейтинга PC;

```python
xone = df[df['platform'] == 'xone']['user_score']
pc = df[df['platform'] == 'pc']['user_score']

alpha = 0.05
results = st.ttest_ind(xone, pc, alternative='greater')
print('p-value:', round(results.pvalue*100, 2), '%')

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: средний пользовательский рейтинг на XOne больше, чем на PC.')
else:
    print('Нет оснований отвергнуть нулевую гипотезу: средние пользовательские рейтинги на XOne и PC равны.')
```

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

### Часть 2

Используем гипотезу о равенстве средних двух генеральных совокупностей, проверяем с помощью st.ttest_ind( )

H0: средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports;

H1: средний пользовательский рейтинг жанра Action больше среднего пользовательского рейтинга жанра Sports;

```python
action = df[df['genre'] == 'Action']['user_score']
sports = df[df['genre'] == 'Sports']['user_score']

alpha = 0.05
results = st.ttest_ind(action, sports, alternative='greater')
print('p-value:', round(results.pvalue*100, 2), '%')

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: средний пользовательский рейтинг жанра Action больше рейтинга Sports.')
else:
    print('Нет оснований отвергнуть нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports равны.')
```

### Пояснение

Нулевая нулевые гипотезы были сформированы исходя из поставленной задачи и применяемого метода для проверки. st.ttest_ind проверяет равенство средних. Альтернативные гипотезы были сформированы с целью получить больше информации, в случае отклонения нулевой гипотезы.
